```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ProdDF = pd.read_csv('products-1000.csv',index_col='Index')
CustDF = pd.read_csv('customers-1000.csv')
```

```python
plt.scatter(ProdDF['Price'],ProdDF['Stock'])
plt.legend()
plt.show()
```

```python
ProdDF['Size'].unique()
col = {'Medium':'green', 'Small':'yellow', 'XXL':'red', 'XL':'orange', 'Extra Large':'black', 'M':'green', 'L':'pink', 'Large':'pink','XS':'purple','NaN':'black'}
plt.scatter(ProdDF['Price'],ProdDF['Stock'],c=ProdDF['Size'].notna.map(col))
plt.show()
```

```python
color_map = ProdDF['Size'].map(col)

# Drop rows where color is NaN (i.e., unrecognized sizes)
filtered_df = ProdDF[color_map.notna()]
color_values = color_map[color_map.notna()].values  # ens
```

```python
plt.scatter(filtered_df['Price'], filtered_df['Stock'], c=color_values)
plt.legend()
plt.show()
```

```python
plt.bar(x = CustDF['Country'],height = CustDF['Country'].count())
```

```python
products = ['Laptops', 'Phones', 'Tablets']
sales = [120, 300, 90]

plt.bar(products,sales, color='skyblue')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales')
plt.show()
```

```python
scores = [55, 67, 72, 73, 68, 90, 88, 75, 82, 60, 59, 95, 77, 62, 83]
print(len(scores))
plt.hist(scores, bins=5, color='lightgreen', edgecolor = 'black')
plt.title('Student Exam score')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

```python
companies = ['Apple', 'Samsung', 'Huawei', 'Others']
shares = [35, 30, 20, 15]
plt.pie(shares,labels = companies,autopct = '%1.1f%%', startangle=190)
plt.show()
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 65, 70, 80, 85, 90]

plt.scatter(study_hours,exam_scores[::-1],color = 'purple')
plt.title("Study Time vs ExamScore")
plt.xlabel("Hour study")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```
